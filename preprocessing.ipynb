```python
#정류소 데이터
spot_df = pd.read_csv("C:/따릉이 데이터셋/3. 공공자전거 대여소 정보(22.06월 기준).csv",encoding='cp949')
```

```python
spot_df = spot_df.dropna(subset=['대여소\n번호','보관소(대여소)명'])
spot_df.isnull().sum()
```

```python
spot_df.columns = ['대여소번호', "대여소명",'위치','상세주소','위도','경도','설치시기','설치형태','거치대수','운영방식']
```

```python
spot_df.head()
```

```python
df.head()
```

```python
df.dtypes
df['대여소번호']=df['대여소번호'].str.replace('"','')
df = df.astype({'대여소번호':'int'})
df.dtypes
```

```python
spot_df = spot_df.astype({'대여소번호':'int'})
spot_df.dtypes
```

```python
df_final = pd.merge(df,spot_df, how='right',on='대여소번호')
df_final.head()
```

```python
del df_final['설치시기']
del df_final['설치형태']
del df_final['거치대수']
del df_final['대여소명_y']
del df_final['운영방식']
```

```python
df_final.to_csv('cp1_df2.csv')
```
