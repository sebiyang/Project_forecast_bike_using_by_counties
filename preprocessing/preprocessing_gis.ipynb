```python
import pandas as pd 

spatial_join= pd.read_csv(r'C:\Users\sebiy\cp1\preprocessed data\따릉이 위치데이터 공간조인.csv',encoding='cp949')

df= pd.read_csv(r'C:\Users\sebiy\cp1\preprocessed data\cp1_df_final_real.csv')

code =pd.read_csv(r'C:\Users\sebiy\cp1\raw dataset\국토교통부_전국 법정동코드.csv',encoding='cp949')
```

```python
spatial_join_copy = spatial_join.copy()
```

```python
spatial_join_copy['EMD_CD'].dropna(inplace=True)
spatial_join_copy['EMD_CD'].fillna(0 , inplace=True)
```

```python
spatial_join_copy.head()
```

```python
spatial_join_copy['EMD_CD'] = spatial_join_copy['EMD_CD'].str.replace('"','')
spatial_join_copy['EMD_CD'] = spatial_join_copy['EMD_CD'].str.replace(' ','')

spatial_join_copy['EMD_CD'] = pd.to_numeric(spatial_join_copy['EMD_CD'],errors='coerce')
#spatial_join_copy = spatial_join_copy.astype({'EMD_CD': 'float'})
#spatial_join_copy = spatial_join_copy.astype({'EMD_CD': 'int'})
```

```python
spatial_join_copy['EMD_CD'] = spatial_join_copy['EMD_CD'].fillna(0).astype('int')
```

```python
spatial_join_copy.head(3)
spatial_join_copy.info()
```

```python
code.head(3)
```

```python
#법정동코드 형식 통일 
spatial_join_copy['EMD_CD'] = spatial_join_copy['EMD_CD'] * 100
```

```python
spatial_join_copy.head(3)
```

```python
#법정동코드 - 법정동명 join

spatial_join_copy.rename(columns={'EMD_CD':'id'},inplace=True)
code.rename(columns={'법정동코드':'id'},inplace=True)
```

```python
spatial_join_fin = pd.merge(spatial_join_copy, code, how='left',on='id')
spatial_join_fin = spatial_join_fin.iloc[:,2:-5]
spatial_join_fin.head(1)
```

```python
df.rename(columns={'대여소번호':'num'},inplace=True)
df_join = pd.merge(spatial_join_fin, df, how='left',on='num')
df_join.head(1)
```

```python
df_join.columns
```

```python
df_join= df_join[['num', 'name', 'loc1', 'loc2', 'lat', 'long','id', '시도명',
       '시군구명', '읍면동명','대여일자', '대여시간',
       '대여구분코드', '성별', '연령대코드', '이용건수', '운동량', '탄소량', '이동거리(M)', '이용시간(분)']]
```

```python
df_join.head(1)
```

```python
df_songpa = df_join[df_join['loc1'] == '송파구']
df_songpa = df_songpa.groupby(['읍면동명','대여일자']).sum()
df_songpa.head(1)
```

```python
df_songpa.reset_index(inplace=True)
df_songpa.head()
```

```python
pt_df_songpa = pd.pivot_table(df_songpa, index='대여일자', columns='읍면동명', values='이용건수')
pt_df_songpa = pt_df_songpa.fillna(0)
pt_df_songpa.tail(2)
```

```python
#나머지 두 개의 구도 동일하게 진행
#강남구

df_gangnam = df_join[df_join['loc1'] == '강남구']
df_gangnam = df_gangnam.groupby(['읍면동명','대여일자']).sum()
df_gangnam.reset_index(inplace=True)
pt_df_gangnam = pd.pivot_table(df_gangnam, index='대여일자', columns='읍면동명', values='이용건수')
pt_df_gangnam = pt_df_gangnam.fillna(0)
pt_df_gangnam.tail(2)
pt_df_gangnam.to_csv('df_gangnam.csv')
```

```python
#영등포구

df_yeong = df_join[df_join['loc1'] == '영등포구']
df_yeong = df_yeong.groupby(['읍면동명','대여일자']).sum()
df_yeong.reset_index(inplace=True)
pt_df_yeong = pd.pivot_table(df_yeong, index='대여일자', columns='읍면동명', values='이용건수')
pt_df_yeong = pt_df_yeong.fillna(0)
pt_df_yeong.tail(2)
pt_df_yeong.to_csv('df_yeong.csv')
```
